```python
import sympy as sm
from sympy import *
import sympy.physics.mechanics as me
me.init_vprinting()
x1,x2 = me.dynamicsymbols('x_1 x_2')
```

```python
eq1 = 0.4*x2 - 1.2*x1 
eq2 = 75*x1 - 25*x2
```

```python
eq1
```

```python
eq2
```

```python
EQ1 = sm.Eq(0,eq1)
EQ2 = sm.Eq(0,eq2)
```

```python
EQ1
```

```python
EQ2
```

```python
points = sm.solve([EQ1,EQ2],x1,x2)
points
```

```python
X = sm.Matrix([[eq1],[eq2]])
Y = sm.Matrix([x1,x2])
J = X.jacobian(Y)
X
```

```python
Y
```

```python
J
```

```python
eig = J.eigenvals()
eig
```

```python
import numpy as np
import sympy as sm
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

w = 10
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
U = 0.4*Y - 1.2*X
V = 75*X - 25*Y
speed = np.sqrt(U**2 + V**2)

fig = plt.figure(figsize=(20, 20))
gs = gridspec.GridSpec(nrows=3, ncols=3, height_ratios=[1, 1, 1])

#  Plot
ax0 = fig.add_subplot(gs[0, 0])
ax0.streamplot(X, Y, U, V, density=[1, 1])
ax0.set_title('Phase Portrait')

plt.tight_layout()
plt.xlim([-10, 10])
plt.ylim([-10, 10])
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
